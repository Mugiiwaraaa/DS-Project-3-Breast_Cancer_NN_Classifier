## Neural_Network - Breast Cancer Classification

```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
import sklearn.datasets
from sklearn.model_selection import train_test_split
```

## Loading the Datasets

```python
br_dataset = sklearn.datasets.load_breast_cancer()
```

```python
br_df=pd.DataFrame(br_dataset.data,columns=br_dataset.feature_names)
```

```python
#Loading the label coloumn
br_df['label']=br_dataset.target
```

## Now splitting the datasets into training and testing data

```python
train_set,test_set=train_test_split(br_df,test_size=0.2,random_state=42)
#80% training data
#20% testing data
```

```python
train_set
```

```python
test_set
```

## Stratified sampling 
to make sure we have similar ratios of features and labels in both testing and training datasets

```python
from sklearn.model_selection import StratifiedShuffleSplit
split =StratifiedShuffleSplit(n_splits=1,test_size=0.2,random_state=42)
for train_index,test_index in split.split(br_df,br_df['label']):
    strat_train_set=br_df.loc[train_index]
    strat_test_set=br_df.loc[test_index]
    
```

```python
strat_test_set
```

```python
strat_train_set
```

## Seperating features and label

```python
X_train = strat_train_set.drop(columns='label',axis=1)
Y_train= strat_train_set['label'].copy()
```

```python
X_test = strat_test_set.drop(columns='label',axis=1)
Y_test= strat_test_set['label'].copy()
```

## Finding correlations

```python
corr_matrix=br_df.corr()
```

```python
corr_matrix['label'].sort_values(ascending=False)
```

## Feature Scaling

```python
# Scaling the features 
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_train_std=scaler.fit_transform(X_train)
X_test_std=scaler.transform(X_test)
```

```python
X_train_std
```

## Building the Neural Network

```python
#Importing Tensorflow and keras for building the neural network 
import tensorflow as tf
tf.random.set_seed(42)
from tensorflow import keras
```

```python
# Setting up layers for the neural network 
model = keras.Sequential([keras.layers.Flatten(input_shape=(30,)),
                          keras.layers.Dense(20,activation='relu'),
                          keras.layers.Dense(2,activation='sigmoid')])
```

```python
#Compiling the neural network 
model.compile(optimizer='adam',
              loss = 'sparse_categorical_crossentropy',
              metrics=['accuracy'])
```

```python
#training the model 
history=model.fit(X_train_std,Y_train,validation_split=0.1,epochs=10)
```

## Visualizing Accuracy and Loss 

```python
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
plt.title('model accuracy ')
plt.ylabel("accuracy")
plt.xlabel("epoch")
plt.legend(['training data','validation data'],loc='lower right')
```

```python
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('model loss ')
plt.ylabel("loss")
plt.xlabel("epoch")
plt.legend(['training data','validation data'],loc='lower right')
```

## Testing the model on the test data 

```python
# evaluate function compares the labels from the test features to the actual test labels and gives us the loss and accuracy of the model
loss,accuracy = model.evaluate(X_test_std,Y_test)
```

## Defining a predictor function 

```python
def Tumor_Predictor(input_data):
    #change input data to numpy array 
    input_data_nparray=np.asarray(input_data)
    #reshaping it as we are predicting only one data point 
    input_data_reshaped=input_data_nparray.reshape(1,-1)
    #Standardizing the input data
    input_data_Std=scaler.transform(input_data_reshaped)
    prediction = model.predict(input_data_Std)
    prediction_label=[np.argmax(prediction)]
    print(prediction_label)
    
    if (prediction_label[0]==0):
        print("The tumor is Malignant.")
    else:
        print("The tumor is Benign.")
            
```
